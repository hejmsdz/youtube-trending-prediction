```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbs
from IPython.display import set_matplotlib_formats

from loaders import load_gb_videos, load_us_videos

%matplotlib inline
set_matplotlib_formats('svg')
```

# Data at the first glance

```python
gb_vids = load_gb_videos()
gb_vids
```

```python
us_vids = load_us_videos()
us_vids
```

```python
# let's merge these dataframes
          
gb_vids['is_GB'] = True
gb_vids['is_US'] = False

us_vids['is_GB'] = False
us_vids['is_US'] = True

vids = pd.concat([gb_vids, us_vids])
#del gb_vids
#del us_vids
print(vids.shape)

vids.head()
# dups = vids.duplicated(vids.columns[:-2])
# print(dups.shape)
# dups.head()
```

# Examining individual attributes

## video_id

```python
# seems quite obvious, but is it really?
vids['video_id'].value_counts()

# apparently videos can show up on the list multiple times
```

```python
# #NAZWA? something is missing and that's a lot of rows!
missing_ids = vids['video_id'] == '#NAZWA?'
vids[missing_ids]
```

```python
# but you know what? we can figure them out from thumbnail URLs
ids = vids[missing_ids]['thumbnail_link'].apply(lambda t: t[23:34])
vids.loc[missing_ids, 'video_id'] = ids
ids
```

```python
# any other suspicious values?
pattern = r'[A-Za-z0-9\-_]{11}'
valid_ids = vids['video_id'].str.match(pattern)
vids[~valid_ids]['video_id']

# yay!
```

```python
# but do you remember that same video IDs appear multiple times?
# let's dig into that again, after filling in missing values

unique_vids = vids['video_id'].nunique()
print(unique_vids)
len(vids) / unique_vids
```

```python
# on average, one video appears around 9 times
# what does the real distribution look like?

count = vids[vids['is_GB']]['video_id'].value_counts()
plt.figure(figsize=(10, 4))
plt.title('is_GB')
ax = sbs.countplot(count)
ax.set(xlabel='Number of occurrences of a single video ID')
plt.show()
print(count.head(5))

count = vids[vids['is_US']]['video_id'].value_counts()
plt.figure(figsize=(10, 4))
plt.title('is_US')
ax = sbs.countplot(count)
ax.set(xlabel='Number of occurrences of a single video ID')
plt.show()
print(count.head(5))
```

```python
# plot shapes are similar for both countries, but there's a difference in scale
# in Great Britain most videos appeared in trending category up to 13 times,
# whereas in the US there's a drop after 7 occurences.
```

```python
gb_ids = set(vids[vids['is_GB']]['video_id'])
us_ids = set(vids[vids['is_US']]['video_id'])
len(gb_ids & us_ids)/len(gb_ids | us_ids)
# almost 12% of videos appeared were trending in both countries
```

```python
# fair enough, but what do multiple occurrences of one video actually mean?
vids[vids['video_id'] == 'j4KvrAUjn6c']

# turns out it could be trending for a longer period and each row corresponds to one day
# number of views, likes, dislikes and comments depends on the date
```

```python
# nevertheless, some (video_id, trending_date) pairs are duplicated
# is it frequent?

scope = ['video_id', 'trending_date']
counts = vids.groupby(scope).size()
counts[counts > 1]
```

```python
# some videos appears both in GB and US at the same time

vids.groupby(scope)\
    .head()\
    .reset_index()\
    .set_index(['video_id', 'trending_date'])\
    .loc['-5aaJJQFvOg', '18.24.02']
```

```python
duplicates = vids.duplicated(vids.columns[:-2], keep=False)
print(duplicates.value_counts(dropna=False))
vids['duplicate'] = duplicates

def markCountries(x):
    if x['duplicate']:
        x['is_GB'] = True
        x['is_US'] = True
    return x

vids = vids.apply(markCountries, axis=1)
print(vids.shape)

vids.drop_duplicates(scope, inplace=True)
vids.drop(columns=['duplicate'], inplace=True)
print(vids.shape)
```

```python
# duplicated videos has been removed

scope = ['video_id', 'trending_date']
counts = vids.groupby(scope).size()
counts[counts > 0]
```

# Extracting time profile

```python
scope = ['video_id', 'trending_date', 'publish_time']
time_profile = vids.groupby(scope)\
    .size()\
    .reset_index()

time_profile.drop(columns=[0], inplace=True)
time_profile
```

```python
# how long does the video stay in Trending?
video_to_days = time_profile.groupby(['video_id']).size()

print(video_to_days.describe())
sbs.distplot(video_to_days, kde=False)

multiple_day_videos = video_to_days.where(video_to_days > 1).count()
print("Videos in 'Trending' for multiple days: " + str(multiple_day_videos))
```

```python
# keep single video row (first mention in Trending) in vids
vids = vids.groupby(['video_id'])\
    .first()\
    .reset_index()

vids
```

```python
# how much time does the video need to become trending?

time_to_trending = pd.to_datetime(vids['trending_date'], utc=True) - pd.to_datetime(vids['publish_time'], utc=True)
print(time_to_trending.describe())

# -1 days - we need to fix the timezones :(
```
